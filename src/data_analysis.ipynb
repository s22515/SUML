Imports + data

```python
from download import import_data_module
import matplotlib.pyplot as plt

data = import_data_module.import_dataframe_from_csv("../model_data/CrabAgePrediction.csv")

original_rows_number = len(data)
```

Check data types

```python
print(data.dtypes)
```

Check if the data is null

```python
missingVals = data.isnull().sum()
print(missingVals)
```

Check for odd data

```python
plots_per_row = 3
plots_number = len(data.columns)
plots_row_number = plots_number // plots_per_row

fig, axs = plt.subplots(plots_row_number, plots_per_row, figsize=(20, 2*plots_number))

for i, col in enumerate(data.columns):
    row = i // plots_per_row
    col_in_row = i % plots_per_row
    axs[row, col_in_row].scatter(data[col], data[col])
    axs[row, col_in_row].set_xlabel(col)
    axs[row, col_in_row].set_ylabel(col)

for i in range(plots_number, plots_row_number):
    axs.flatten()[i].axis('off')

plt.show
```

Deleting odd data

-- Reset data

```python
data = import_data_module.import_dataframe_from_csv("../model_data/CrabAgePrediction.csv")
```

-- Delete data

```python
data = data[data['Viscera Weight'] < 17.5]
data = data[data['Shell Weight'] <= 24]
data = data[data['Shucked Weight'] <= 36]
data = data[data['Weight'] <= 78]
data = data[data['Height'] < 1]
```

Review data after clening odd data

```python
plots_per_row = 3
plots_number = len(data.columns)
plots_row_number = plots_number // plots_per_row

fig, axs = plt.subplots(plots_row_number, plots_per_row, figsize=(20, 2*plots_number))

for i, col in enumerate(data.columns):
    row = i // plots_per_row
    col_in_row = i % plots_per_row
    axs[row, col_in_row].scatter(data[col], data[col])
    axs[row, col_in_row].set_xlabel(col)
    axs[row, col_in_row].set_ylabel(col)

for i in range(plots_number, plots_row_number):
    axs.flatten()[i].axis('off')

plt.show
```
